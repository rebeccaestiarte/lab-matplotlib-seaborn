#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# Creating 2 subplots
fig, (ax1, ax2) = plt.subplots(1,2, figsize = (13,5))
fig.suptitle('Graphs 2', fontsize = 20)

# definining subplot 1
ax1.plot(x, x**2, "orangered", linewidth = 4)
ax1.set_xlabel("x", fontsize = 15)
ax1.set_ylabel("x^2", fontsize = 15)
ax1.set_xscale("linear")

# definining subplot 2
y = np.exp(x)

ax2.plot(x, y, "deepskyblue", linewidth = 4)
ax2.set_xlabel("x", fontsize = 15)
ax2.set_ylabel("e^x", fontsize = 15)
ax2.set_yscale("log")
ax2.set_xscale("log")

plt.show() 
plt.clf()
```

#### As a bonus challenge, try to add a legend to the plot.

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize = (13,5))
fig.suptitle('Graphs 2', fontsize = 20)

ax1.plot(x, x**2, "orangered", label = "y=x^2", linewidth = 4)
ax1.set_xlabel("x", fontsize = 15)
ax1.set_ylabel("x^2", fontsize = 15)
ax1.set_xscale("linear")
ax1.legend()

# definining subplot 2
y = np.exp(x)

ax2.plot(x, y, "deepskyblue", label = "y=e^x", linewidth = 4)
ax2.set_xlabel("x", fontsize = 15)
ax2.set_ylabel("e^x", fontsize = 15)
ax2.set_yscale("log")
ax2.set_xscale("log")
ax2.legend()

plt.show() 
plt.clf()
```
